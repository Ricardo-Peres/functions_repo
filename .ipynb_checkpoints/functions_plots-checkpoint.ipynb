### General plots

```python
def plot_cs2cs1(df, df_cut, bins=200,range=[[0,200],[0,3.5]],save_fig = False,tag=''):
    cmap = 'winter'
    
    plt.figure(figsize=(15,5))
    plt.subplot(121)
    plt.hist2d(df['cs1'], np.log10(df['cs2_bottom'] / df['cs1']), 
                bins=bins, norm=LogNorm(), range = [[0,250],[-2,9]], cmap=cmap)


    plt.xlabel('cS1')
    plt.ylabel('log10(cS2b / cS1)')
    plt.title('No cuts')
    plt.colorbar()


    plt.subplot(122)
    plt.hist2d(df_cut['cs1'], np.log10(df_cut['cs2_bottom'] / df_cut['cs1']), 
                bins=bins, norm=LogNorm(), range = range, cmap=cmap)


    plt.xlabel('cS1')
    plt.ylabel('log10(cS2b / cS1)')
    plt.title('With cuts')
    plt.colorbar()
    
    plt.suptitle('%s' %df.name, fontsize=20)

    if save_fig == True:
        plt.savefig('./figures/%scs2cs1_%s.png' %(tag,df.name), bbox_inches='tight')
        print('Figure saved')
        plt.show()
    elif save_fig == False:
        plt.show()
```

```python
def plot_xyrzplot(df,df_cut,save_fig=False,tag=''):
    cmap = matplotlib.cm.winter
    lwidth = 0.8
    colorTPC = 'red'
    
    fig = plt.figure(figsize=(12,10))
    
    ax1 = fig.add_subplot(221)

    ax1.add_patch(Circle((0,0), tpc_radius, color=colorTPC, fill = False, linewidth=lwidth) )

    #plt.title('No cuts')

    hist1=plt.hist2d(df['x_3d_nn'], df['y_3d_nn'], bins = 100, cmin = 1,cmap = cmap,norm=LogNorm())

    plt.xlim((-60,60))
    plt.ylim((-60,60))
    plt.xlabel('x_3d_nn (cm)')
    plt.ylabel('y_3d_nn (cm)')
    #plt.colorbar(label = 'events')
    
    plt.gca().set_aspect('equal', adjustable='box')



    ax2 = fig.add_subplot(223)

    ax2.add_patch(Rectangle((0,-tpc_length),tpc_radius, tpc_length, color=colorTPC, fill = False, linewidth=lwidth) )

    #plt.title('No cuts')

    hist2 = plt.hist2d(df['r_3d_nn'], df['z_3d_nn'], bins = 100, cmin = 1, range = [[0,60],[-150,10]],\
               cmap = cmap,norm=LogNorm())

    plt.xlabel('r_3d_nn (cm)')
    plt.ylabel('z_3d_nn (cm)')
    #plt.colorbar(label = 'events')
    #plt.clim(1,np.nanmax(np.array([np.ravel(hist1[0]),np.ravel(hist2[0])])))
    #plt.gca().set_aspect('equal', adjustable='box')

    
    ax3 = fig.add_subplot(222)

    ax3.add_patch(Circle((0,0), tpc_radius, color=colorTPC, fill = False, linewidth=lwidth) )

    #plt.title('With cuts')

    hist3 = plt.hist2d(df_cut['x_3d_nn'], df_cut['y_3d_nn'], bins = 100, cmin = 1,cmap = cmap,norm=LogNorm())

    plt.xlim((-60,60))
    plt.ylim((-60,60))
    plt.xlabel('x_3d_nn (cm)')
    plt.ylabel('y_3d_nn (cm)')
    plt.colorbar(label = 'events')
    #plt.clim(1,100)
    plt.gca().set_aspect( 'auto',adjustable='box', anchor = 'SW')
    


    ax4 = fig.add_subplot(224)

    ax4.add_patch(Rectangle((0,-tpc_length),tpc_radius, tpc_length, color=colorTPC, fill = False, linewidth=lwidth) )

    #plt.title('With cuts')fit

    hist4 = plt.hist2d(df_cut['r_3d_nn'], df_cut['z_3d_nn'], bins = 100, cmin = 1, range = [[0,60],[-150,10]],\
               cmap = cmap,norm=LogNorm())

    plt.xlabel('r_3d_nn (cm)')
    plt.ylabel('z_3d_nn (cm)')
    plt.colorbar(label = 'events')
    #plt.gca().set_aspect('equal', adjustable='box')

    #plt.clim(1,100)
    plt.tight_layout()
    plt.suptitle('%s' %df.name, fontsize=20)
    
    if save_fig == True:
        plt.savefig('./figures/%sspatial_distribution_%s.png' %(tag,df.name), bbox_inches='tight')
        print('Figure saved')
        plt.show()
    elif save_fig == False:
        plt.show()
```

### Leakage analysis plots

```python
def plot_leakdif_cs2cs1plot(df_cut, bins=200,save_fig = False):
    cmap = 'winter'
    
    plt.figure(figsize=(15,5))
    plt.subplot(121)
    plt.hist2d(df_cut['cs1'], np.log10(df_cut['cs2_bottom'] / df_cut['cs1']), 
                bins=bins, norm=LogNorm(), range = [[0,250],[0,3.5]], cmap=cmap)


    plt.xlabel('cS1')
    plt.ylabel('log10(cS2b / cS1)')
    plt.title('With leakage')
    plt.colorbar()


    plt.subplot(122)
    plt.hist2d(df_cut[df_cut['is_leakage']==False]['cs1'], 
               np.log10(df_cut[df_cut['is_leakage']==False]['cs2_bottom'] / df_cut[df_cut['is_leakage']==False]['cs1']), 
                bins=bins, norm=LogNorm(), range = [[0,200],[0,3.5]], cmap=cmap)


    plt.xlabel('cS1')
    plt.ylabel('log10(cS2b / cS1)')
    plt.title('Without leakage')
    plt.colorbar()
    
    plt.suptitle('All pos', fontsize=20)

    if save_fig == True:
        plt.savefig('./figures/cs2cs1_leak.png', bbox_inches='tight')
        print('Figure saved')
        plt.show()
    elif save_fig == False:
        plt.show()
```

```python
def plot_leakdtscutevents(df_cut,dtscut,save_fig=False):
    plt.figure(figsize = (10,6))
    plt.hist2d(df_cut['cs1'], np.log10(df_cut['cs2_bottom'] / df_cut['cs1']), 
                    bins=100, norm=LogNorm(), range = [[0,200],[0,3.5]], cmap='winter')
    plt.xlabel('cS1')
    plt.ylabel('log10(cS2 / cS1)')
    plt.colorbar()
    dtscut = 83
    plt.scatter(df_cut[df_cut['dts']>dtscut]['cs1'], 
                np.log10(df_cut[df_cut['dts']>dtscut]['cs2_bottom'] / df_cut[df_cut['dts']>dtscut]['cs1']),
                marker = 'x', color = 'red', s = 23)

    if save_fig == True:
        plt.savefig('./figures/leakage_dts_cs2cs1.png' , bbox_inches='tight')
        print('Figure saved')
        plt.show()
    elif save_fig == False:
        plt.show()
```

```python
def plot_perc_bands(df_cut,fit_values, bins=100,function = 'SR0',save_fig = False, percentiles= False,tag = ''):
    cmap = 'winter'
    if percentiles!= False:
        alpha = 0.7
    else:
        alpha = 0.7
    plt.figure(figsize=(12,6))
    
    plt.hist2d(df_cut['cs1'], np.log10(df_cut['cs2_bottom'] / df_cut['cs1']), 
                bins=bins, norm=LogNorm(), range = [[1,150],[0,3]], cmap=cmap, alpha =alpha)
    plt.colorbar()
    if function == 'SR1':
        #5th/95th percentile
        x = np.linspace(0.1,150,1000)

        plt.plot(x,fit_function_SR1(x,fit_values[0][0],fit_values[0][1],fit_values[0][2],fit_values[0][3],fit_values[0][4]),'r-',lw=3,label= '5th/95th percentile')
        plt.plot(x,fit_function_SR1(x,fit_values[2][0],fit_values[2][1],fit_values[2][2],fit_values[2][3],fit_values[2][4]),'r-',lw=3)

        #Median
        plt.plot(x,fit_function_SR1(x,fit_values[1][0],fit_values[1][1],fit_values[1][2],fit_values[1][3],fit_values[1][4]),'r--',lw=3,label= 'median')
    
    elif function == 'SR0':
        #5th/95th percentile
        x = np.linspace(0.1,150,bins)

        plt.plot(x,fit_function_SR0(x,fit_values[0][0],fit_values[0][1],fit_values[0][2],fit_values[0][3]),'r--',lw=3,label= '5th/95th percentile')
        plt.plot(x,fit_function_SR0(x,fit_values[2][0],fit_values[2][1],fit_values[2][2],fit_values[2][3]),'r--',lw=3)

        #Median
        plt.plot(x,fit_function_SR0(x,fit_values[1][0],fit_values[1][1],fit_values[1][2],fit_values[1][3]),'r-',lw=3,label= 'median')
    
    else:
        #5th/95th percentile
        x = np.linspace(0.1,150,bins)

        plt.plot(x,function(x,fit_values),'r-',lw=3,label= '5th/95th percentile')
        plt.plot(x,function(x,fit_values),'r-',lw=3)

        #Median
        plt.plot(x,function(x,fit_values),'--',color='black',lw=3,label= 'median')
    
    
    plt.xlabel('cS1')
    plt.ylabel('log10(cS2b / cS1)')
    plt.title('Fitted Band')
    
    
    plt.legend(fontsize = 13,frameon=False)
    #lt.suptitle('%s' %df.name, fontsize=20)
    
    if save_fig == True:
        plt.savefig('./figures/%sfittedband_%s.png' %(tag,df_cut.name) , bbox_inches='tight')
        print('Figure saved')
        plt.show()
    elif save_fig == False:
        plt.show()
```

```python
def plot_bandcompare(df_cut, fit_values, bins=100,save_fig = False,usrlabel=''):
    from matplotlib import gridspec
    fig = plt.figure(figsize=[10,10])
    gs = gridspec.GridSpec(2,1,height_ratios=[5,1])
    ax0 = plt.subplot(gs[0])
    plt.hist2d(df_cut['cs1'], np.log10(df_cut['cs2_bottom'] / df_cut['cs1']), 
                    bins=bins,cmin = 1,norm=LogNorm(), range = [[1,150],[0.5,3]], cmap='winter', alpha =0.2)
    
    x=np.linspace(2,200,1000)
    fit5 = fit_function_SR0(x,fit_values[0][0],fit_values[0][1],fit_values[0][2],fit_values[0][3])#,fit_values[0][4])
    fit95 = fit_function_SR0(x,fit_values[2][0],fit_values[2][1],fit_values[2][2],fit_values[2][3])#,fit_values[2][4])
    fit50 = fit_function_SR0(x,fit_values[1][0],fit_values[1][1],fit_values[1][2],fit_values[1][3])#,fit_values[1][4])
    NR_SR1 = fit_function_SR1(x,a_mean_1,b_mean_1,c_mean_1,d_mean_1,e_mean_1)
    NR_SR0 = fit_function_SR1(x,a_mean_0,b_mean_0,c_mean_0,d_mean_0,e_mean_0)
    #SR1 & SR0 medians
    #plt.fill_between(x,fit_function_SR1(x,a_high,b_high,c_high,d_high,e_high), 
    #                 fit_function_SR1(x,a_low,b_low,c_low,d_low,e_low),
    #                 alpha=0.5,color = 'cyan')
    plt.plot(x, NR_SR0,'grey',lw = 2,label='SR0 NR median (AmBe)')

    plt.plot(x, NR_SR1,'black',lw = 2,label='SR1 NR median (NG + AmBe)')
    
    #This work
    plt.fill_between(x,fit95, 
                     fit5,
                     alpha=0.5,color = 'orange')
    plt.plot(x,fit50,'red', lw = 2, label = 'This work')
    plt.ylabel('log10(cS2_b / cS1)')
    #plt.title('NR Band compare')
    plt.legend(fontsize  = 13,frameon=False)
    
    
    ax1 = plt.subplot(gs[1],sharex=ax0)
    x=np.linspace(2,150,75)
    fit50 = fit_function_SR0(x,fit_values[1][0],fit_values[1][1],fit_values[1][2],fit_values[1][3])#,fit_values[1][4])
    NR_SR1 = fit_function_SR1(x,a_mean_1,b_mean_1,c_mean_1,d_mean_1,e_mean_1)
    NR_SR0 = fit_function_SR1(x,a_mean_0,b_mean_0,c_mean_0,d_mean_0,e_mean_0)
    
    plt.setp(ax0.get_xticklabels(), visible=False)
    plt.plot(x, (fit50-NR_SR0)/NR_SR0*100,'grey',ls = '', marker = '.',lw = 3, label= 'relative difference SR0')
    plt.plot(x, (fit50-NR_SR1)/NR_SR1*100,'black',ls = '', marker = '.',lw = 3, label= 'relative difference SR1')
    plt.axhline(y=0, color= 'grey', ls= '--',alpha = 0.7)

    #plt.ylim(-0.05*100,0.05*100) #comment to allow free y on relative difference
    #plt.ylabel('(NG_2018-NR_SR1)/NR_SR1')
    plt.ylabel('Relative difference (%)')
    plt.xlabel('cS1')
    plt.xlim(1,150)
    plt.legend(fontsize  = 10,frameon=False)
    if save_fig == True:
        plt.savefig('./figures/%sbandcompare_%s.png'%(usrlabel,df_cut.name), bbox_inches='tight')
        print('Figure saved')
        plt.show()
    elif save_fig == False:
        plt.show()
```

```python
def plot_gauss_bands(df_cut,fit_values,xlog=False, bins=200,function = 'SR0', df_gauss= None,save_fig = False,tag = ''):
    cmap = 'winter'
    alpha = 0.4
    plt.figure(figsize=(12,6))
    
    plt.hist2d(df_cut['cs1'], np.log10(df_cut['cs2_bottom'] / df_cut['cs1']), 
                bins=bins, norm=LogNorm(), range = [[1,150],[0,3]], cmap=cmap, alpha =alpha)
    plt.colorbar()
    
    if function == 'SR1':
        #5th/95th percentile
        x = np.linspace(0.1,150,1000)
        a,b,c,d,e = fit_values['p2sig'][0],fit_values['p2sig'][0][1],fit_values['p2sig'][0][2],fit_values['p2sig'][0][3],fit_values['p2sig'][0][4]
        plt.plot(x,fit_function_SR1(x,a,b,c,d,e),'r-',lw=3,label= '5th/95th percentile')
        a,b,c,d,e = fit_values['m2sig'][0][0],fit_values['m2sig'][0][1],fit_values['m2sig'][0][2],fit_values['m2sig'][0][3],fit_values['m2sig'][0][4]
        plt.plot(x,fit_function_SR1(x,a,b,c,d,e),'r-',lw=3)

        #Median
        a,b,c,d,e = fit_values['mu'][0][0],fit_values['mu'][0][1],fit_values['mu'][0][2],fit_values['mu'][0][3],fit_values['mu'][0][4]
        plt.plot(x,fit_function_SR1(x,a,b,c,d,e),'r--',lw=3,label= 'median')
    
    elif function == 'SR0':
        #5th/95th percentile
        x = np.linspace(0.1,150,bins)
        a,b,c,d = fit_values['p2sig'][0][0],fit_values['p2sig'][0][1],fit_values['p2sig'][0][2],fit_values['p2sig'][0][3]
        plt.plot(x,fit_function_SR0(x,a,b,c,d),'r-',lw=3,label= '5th/95th percentile')
        
        a,b,c,d = fit_values['p2sig'][0][0],fit_values['m2sig'][0][1],fit_values['m2sig'][0][2],fit_values['m2sig'][0][3]
        plt.plot(x,fit_function_SR0(x,a,b,c,d),'r-',lw=3)

        #Median
        a,b,c,d = fit_values['mu'][0][0],fit_values['mu'][0][1],fit_values['mu'][0][2],fit_values['mu'][0][3]

        plt.plot(x,fit_function_SR0(x,a,b,c,d),'r--',lw=3,label= 'median')
    
    else:
        #5th/95th percentile
        x = np.linspace(0.1,150,bins)

        plt.plot(x,function(x,fit_values),'r-',lw=3,label= '5th/95th percentile')
        plt.plot(x,function(x,fit_values),'r-',lw=3)

        #Median
        plt.plot(x,function(x,fit_values),'--',color='black',lw=3,label= 'median')
    
    if type(df_gauss) == pd.core.frame.DataFrame:

        plt.plot(df_gauss['x_mid'], df_gauss['mu'], marker = '*',color = 'black' ,label='median', linestyle='')
        plt.plot(df_gauss['x_mid'], df_gauss['mu-2sig'],'+',color = 'black', label='-2 sigma',  linestyle='')
        plt.plot(df_gauss['x_mid'], df_gauss['mu+2sig'],'x', color = 'black',label = '+2 sigma', linestyle='')
    
    if xlog == True:
        plt.xscale('log')
    plt.xlabel('cS1')
    plt.ylabel('log10(cS2b / cS1)')
    plt.title('Fitted Band')
    
    
    plt.legend(fontsize = 13,frameon=False)
    #lt.suptitle('%s' %df.name, fontsize=20)
    
    if save_fig == True:
        plt.savefig('./figures/%sfittedband_%s.png' %(tag,df_cut.name) , bbox_inches='tight')
        print('Figure saved')
        plt.show()
    elif save_fig == False:
        plt.show()
```
